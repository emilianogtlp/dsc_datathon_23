```python
df = pd.read_csv("/Users/emilianogtlp/Documents/proyectos/dsc_datathon_23/dataset_transacc_tdc_tdd_20230309.csv", parse_dates=['fecha_transaccion'])
```

```python
df['mes'] = df['fecha_transaccion'].dt.month
df['anio'] = df['fecha_transaccion'].dt.year
clients = df[['fecha_transaccion','mes','anio','sexo_cliente','estado_cliente','edad_cliente','monto_transaccion','id_cliente']].groupby(['id_cliente','anio','mes'])['monto_transaccion'].sum().reset_index()
```

```python
clients.head(20)
```

```python
clients.describe()
```

```python
fig = px.box(clients, x="mes",y="monto_transaccion")
fig.show()
```

```python
ids = df["id_cliente"].unique().tolist()
```

```python
mean = 3000
std = 150
salario = np.clip(np.random.normal(mean,std,10000),150,5000)
```

```python
df_salarios = pd.DataFrame({'id_cliente':ids,'salario':salario})
```

```python
merged = pd.merge(clients,df_salarios, on = 'id_cliente')
merged['score'] = merged['monto_transaccion']/merged['salario']
merged
```
